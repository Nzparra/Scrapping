```python
from bs4 import BeautifulSoup
from bs4 import NavigableString
import time
import requests
import json


def obtain_floor_availability():
    url =  'https://www.bellatplantation.com/rentCafeAPIConnect.php?_=1631070176685'
    page = requests.get(url)
    time.sleep(1)
    soup = BeautifulSoup(page.content, 'html.parser').text
    data = json.loads(soup)
    for i in data:
        print("Name:", i['FloorplanName'])
        print("Beds:", i['Beds'])
        print("Baths:", i['Baths'])
        print("Sq. Ft:", i['MaximumSQFT'])
        print("AvailabilityURL:", i['AvailabilityURL'])
        print("MinimumRent:",i['MinimumRent'])
        print("MaximumRent:",i['MaximumRent'])
        print()
        
obtain_floor_availability()
```
