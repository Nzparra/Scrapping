```python
import pandas as pd

def split_number_of_unique_organizations(row):
    organizations = row.split(",")
    clean_organizations = {}
    for organization in organizations:
        if organization in clean_organizations:
            clean_organizations[organization] += 1
        else:
            clean_organizations[organization] = 0
    return len(clean_organizations)

def get_most_repeated_organization(row):
    organizations = row.split(",")
    clean_organizations = {}
    for organization in organizations:
        if organization in clean_organizations:
            clean_organizations[organization] += 1
        else:
            clean_organizations[organization] = 0
    most_repeated = max(clean_organizations, key=clean_organizations.get)
    if(clean_organizations[most_repeated] == 0):
        return "All the Organizations appear only once"
    return most_repeated

df = pd.read_csv('Number of unique values of organizations and most repeated organization.csv')
df['Number of unique organizations'] = df['Organizations'].apply(lambda row: split_number_of_unique_organizations(row))
df['Most repeated organization'] = df['Organizations'].apply(lambda row: get_most_repeated_organization(row))
df.to_excel('Clean_Organizations.xlsx', encoding='utf-8', index=False)
df.head()

```
